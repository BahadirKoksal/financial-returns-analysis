## Veri keşfi

Finansal verileri bir `df` değişkenine yüklemek için aşağıdaki hücreyi çalıştırın.

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv')
df.head()
```

`.info()` yöntemini kullanarak veri türlerine göz atın.

```python
df.info()
```

Uygun işlevi kullanarak **istatistiksel bilgileri** görüntüleyin.

```python
df.describe()
```

```python
df["Date"]=pd.to_datetime(df["Date"])
```

## Yeniden örnekleme ve iadeler

`Date` sütununu indeks olarak ayarlayın.



```python
df = df.set_index("Date")
df.head()
```

**.resample()** işlevini kullanarak, her çeyrek için `AAPL.Open` minimum değerini ve `AAPL.Close` medyan değerini hesaplayın.

<summary> İpucu 🔍 </summary>

## Pencere(Window) işlevleri

`AAPL.Close`'un **son 4 gün** boyunca hareketli ortalamasını hesaplayın.

```python
df['AAPL.Close'].rolling(4).mean()
```

`.rolling()` yöntemi, sadece önceki satırların sayısını değil, bir zaman aralığını da belirlememizi sağlar. \
Bu, ilk satırlardaki NaN değerlerini ortadan kaldırmak için kullanışlıdır.


`AAPL.Close` değerinin hareketli ortalamasını **30 günlük bir kayan pencere kullanarak** hesaplayın.

```python
df["AAPL.Close"].rolling("30D").mean()
```

## Kayıp değerler

Veri setimizde hafta sonu günleri eksik.


Dizin tarihlerindeki boşlukları doldurmak için DataFrame `.asfreq()` yöntemini kullanabiliriz. \
[Belgelere](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.asfreq.html) göz atın ve hafta sonu günlerini içeren `df_with_weekends` adlı yeni bir DataFrame oluşturun.

```python
df_with_weekends = df.asfreq('D')
df_with_weekends.head(10)
```

`asfreq()` işlevi tarafından oluşturulan tüm NaN'leri ortadan kaldırmamız gerekiyor.


`fillna()` yöntemine [belgelere](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html#pandas.DataFrame.fillna) bakın ve bunu kullanarak eksik değerleri **bir sonraki geçerli değerle** değiştirin.


⚠️ Değeri `df_with_weekends`'e yeniden atamayın.

```python
df_with_weekends.fillna(method='bfill')
```

Eksik değerleri değiştirmek için başka bir yöntem de [interpolasyon](https://en.wikipedia.org/wiki/Interpolation) kullanmaktır.


`interpolate()` yöntemine bakın ve eksik değerleri değiştirmek için kullanın. \
`df_with_weekends` değişkeninizi güncelleyebilirsiniz.

```python
df_with_weekends = df_with_weekends.interpolate()
df_with_weekends.head(10)
```

## Financial charts


Grafik oluşturma becerilerimizi geliştirme zamanı! 📈📉


Şu ana kadar, çalışmalarımızı yalnızca Apple hisselerine odakladık. Şimdi, birden fazla şirketin hisselerini de dahil ederek, yatırımcıların kararları için ihtiyaç duydukları temel grafikleri oluşturacağız.


**Python'un finans dünyasında neden bu kadar ün kazandığını keşfetmek üzeresiniz.** 🏦




<u>Not:</u> **Ticker**, belirli bir borsada belirli bir hisseyi belirtmek için kullanılan kısaltmadır (**ör.** Apple için *AAPL*, Tesla için *TSLA* vb.).


---








Birden fazla ticker'dan veri yüklemek için aşağıdaki hücreyi çalıştırın.


Değerlerin Day One'a göre yeniden bazlandığını unutmayın. Diğer bir deyişle, gördüğünüz şey orijinal ticker değeri değil, getiri değeridir.

```python
df = px.data.stocks()

df["date"] = pd.to_datetime(df["date"], format='%Y-%m-%d', errors='coerce')
df = df.set_index("date")
df
```

`GOOG`'un gelişimini çizgi grafiği ile gösterin.

```python
px.line(df, x=df.index, y='GOOG', title='GOOG Gelişimi')
```

GOOG'un minimum ve maksimum kümülatif değeri ile birlikte gelişimini **aynı grafikte** gösterin.

<details>
<summary>İpucu 🔍</summary>

- `cummin()` ve `cummax()` yöntemlerine bir göz atın.
- Plotly grafiğine bir çizgi eklemek için `.add_scatter()` yöntemini kullanabilirsiniz.

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df.index, y=df['GOOG'], mode='lines', name='GOOG'))
fig.add_trace(go.Scatter(x=df.index, y=df['GOOG'].cummax(), mode='lines', name='Kümülatif Maksimum'))
fig.add_trace(go.Scatter(x=df.index, y=df['GOOG'].cummin(), mode='lines', name='Kümülatif Minimum'))
fig.update_layout(title='GOOG Gelişimi ile Kümülatif Min/Maks', xaxis_title='date', yaxis_title='GOOG')
fig.show()
```

GOOG'un gelişimini ve 10 günlük ve 30 günlük hareketli ortalamalarını **aynı grafikte** gösterin.

<details>
<summary>İpucu 🔍</summary>


burada `.rolling()` yöntemini kullanmak isteyebilirsiniz


</details>

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df.index, y=df['GOOG'], mode='lines', name='GOOG'))
fig.add_trace(go.Scatter(x=df.index, y=df['GOOG'].rolling(10).mean(), mode='lines', name='10 Günlük Hareketli Ortalama'))
fig.add_trace(go.Scatter(x=df.index, y=df['GOOG'].rolling(30).mean(), mode='lines', name='30 Günlük Hareketli Ortalama'))
fig.update_layout(title='GOOG Gelişimi ile Hareketli Ortalamalar', xaxis_title='date', yaxis_title='GOOG')
fig.show()
```

___


Şimdi her bir ticker için getirileri ayrı grafiklerde göstereceğiz.

`pd.melt()` işlevi, geniş formatlı bir tabloyu daha uzun ve daha yapılandırılmış bir formata dönüştürebilir. \
Sütunlardan değişkenleri alır ve bunları satırlara dönüştürür, değişken adları için yeni bir sütun ve karşılık gelen değerler için başka bir sütun oluşturur.


Aşağıdaki hücreyi çalıştırın.

```python
df_melt = pd.melt(df, var_name="ticker", ignore_index=False)
df_melt
```

Yeni oluşturulan DataFrame üzerinde `px.area()` işlevini kullanarak,


her bir ticker için **ayrı grafiklerde** getirileri görüntüleyin.

<details>
<summary> İpucu 🔍 </summary>

Ders notlarında çok benzer bir örnek gösterilmişti.


Şimdi `facet_col` komutunun ne işe yaradığını hatırlamanın tam zamanı!


</details>

```python
px.area(df_melt, x=df_melt.index, y='value', facet_col='ticker', facet_col_wrap=3, title='Ticker Bazında Getiriler')
```
